# Data Preprocessing

The following jupyter notebook contains all of the steps done in data preprocessing research. The key difference is that we are using a class created in the utils folder to retrieve the preprocessed dataframe that would be used to perform EDA and classification problems using different ML algorithms. The steps taken are:
1. Drop Unique Identifiers from credit card history `Unnamed:0`, `trans_num`.
2. Fill Nan values in `merch_zipcode` with the median value.
3. Obtain the datetime features from transaction and date of birth datetimes: `trans_date_trans_time`, `dob`
4. Obtain low categorical features, binary and small unique identifiers: `gender`, `category`, `state`
5. Obtain high categorical features, using frequency and target encoding: `merchant`, `first`, `last`, `street`, `city`, `job`
